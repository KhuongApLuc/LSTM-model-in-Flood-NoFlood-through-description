```python
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv('devimg_describe.csv')

# Kiểm tra missing value
print(df.isnull().sum())

# Xử lý missing value bằng cách thay thế bằng chuỗi rỗng
df['description'].fillna('', inplace=True)
df['user_tags'].fillna('', inplace=True)
```

#### Xoá các cột không cần thiết

```python
df = df.drop(columns=['license_name'])
df = df.drop(columns=['user_nsid'])
df = df.drop(columns=['image_extension_original'])
df = df.drop(columns=['longitude'])
df = df.drop(columns=['license_url'])
df = df.drop(columns=['date_uploaded'])
df = df.drop(columns=['date_taken'])
df = df.drop(columns=['latitude'])
df = df.drop(columns=['image_url'])
df = df.drop(columns=['user_nickname'])
df = df.drop(columns=['capture_device'])
df = df.drop(columns=['title'])

df.to_csv('description_id.csv', index=False)
```

#### Ghép file chứa id/label và file chứa id/features (dựa vào id)

```python
df_ = pd.read_csv('devset_images_gt.csv')
merged_df = pd.merge(df, df_, left_on='image_id', right_on='id')
print("Các cột sau khi ghép:", merged_df.columns)
merged_df.to_csv('final_describe.csv', index=False)
```

#### Ghép cột description và user_tags

```python
from sklearn.model_selection import train_test_split

# Kết hợp cột description và tag thành một cột duy nhất
merged_df['text'] = df['description'] + ' ' + merged_df['user_tags']

# Tách dữ liệu thành training và testing set
X_train, X_test, y_train, y_test = train_test_split(merged_df['text'], merged_df['label'], test_size=0.2, random_state=42)
```

```python
merged_df.drop(columns=['user_tags'], inplace=True)
merged_df.drop(columns=['description'],inplace=True)
```

#### LSTM model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, SpatialDropout1D, Bidirectional
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Khởi tạo tokenizer
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(X_train)

# Chuyển văn bản thành sequences
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

# Đệm sequences để có cùng chiều dài
max_len = 1000
X_train_pad = pad_sequences(X_train_seq, maxlen=max_len)
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len)

# Khởi tạo mô hình LSTM
model_ = Sequential()
model_.add(Embedding(input_dim=10000, output_dim=128, input_length=max_len))
model_.add(SpatialDropout1D(0.2))
model_.add(Bidirectional(LSTM(64, dropout=0.2, recurrent_dropout=0.2)))
model_.add(Dense(1, activation='sigmoid'))

# Compile mô hình
model_.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Huấn luyện mô hình
history = model_.fit(X_train_pad, y_train, epochs=10, batch_size=128, validation_data=(X_test_pad, y_test), verbose=1)

# Đánh giá mô hình
loss, accuracy = model_.evaluate(X_test_pad, y_test, verbose=0)
print(f'Accuracy: {accuracy}')
```
